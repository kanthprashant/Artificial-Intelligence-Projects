```python
# Importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Importing the dataset
dataset = pd.read_csv('/content/drive/My Drive/Imitation Game/ag3_5by5dGrid.csv', names = range(27))
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
```

```python
dataset.head()
```

```python
# Encoding categorical data
from sklearn.preprocessing import LabelEncoder
labelencoder_y = LabelEncoder()
y = labelencoder_y.fit_transform(y)
```

```python
# Encoding categorical data
from tensorflow.keras.utils import to_categorical

# make the model and load the training dataset.

y = to_categorical(y, 4)

# call the fit method.
```

```python
# Splitting the dataset into train and test
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.1, random_state =0)
```

```python
# Feature Scaling
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
X_train = sc_X.fit_transform(X_train)
X_test = sc_X.transform(X_test)
```

```python
# For convolution models
#X_train = np.reshape(X_train, (-1, 5, 5, 1))
#X_test = np.reshape(X_test, (-1, 5, 5, 1))
```

```python
# Importing keras libraries
import keras
from keras.models import Sequential
from keras.layers.convolutional import Convolution2D
from keras.layers import MaxPooling2D
from keras.layers import Flatten
from keras.layers import Dense

# Initialising the Classifier
classifier = Sequential()
classifier.add(Dense(units = 26, activation = 'relu'))
#classifier.add(Convolution2D(64, (1, 1), padding="valid", activation="relu", input_shape=(5,5,1)))
#classifier.add(Convolution2D(32, (1, 1), padding="valid", activation="relu", input_shape=(8,8,1)))
#classifier.add(MaxPooling2D(pool_size = (2, 2)))
#classifier.add(Flatten())
#classifier.add(MaxPooling2D(pool_size=(2, 2)))
#classifier.add(Dense(units = 256, activation = 'relu'))
#classifier.add(Dense(units = 128, activation = 'relu'))
#classifier.add(Dense(units = 8, activation = 'relu'))
classifier.add(Dense(units = 16, activation = 'relu'))
classifier.add(Dense(units = 8, activation = 'relu'))
classifier.add(Dense(units = 4, activation = 'softmax'))
```

```python
# Compiling the ANN
from tensorflow.keras.optimizers import Adam
classifier.compile(optimizer = Adam(learning_rate= 0.0025), loss = 'categorical_crossentropy', metrics = ['accuracy'])
```

```python
# Fitting the ANN to training set
classifier.fit(X_train, y_train, batch_size = 128, epochs = 10)
```

```python
# Making predictions and evaluating the model
y_pred = classifier.predict(X_test)
```

```python
import numpy as np
count = 0
total = len(y_pred)
for i in range(total):
  if np.argmax(y_pred[i]) == np.argmax(y_test[i]):
    count += 1
print("Accuracy: ", count/total)
```
